```python
import pandas as pd
import json
import uuid
import time
```

```python
df = pd.read_json("hf://datasets/AdaptLLM/finance-tasks/Headline/test.json")
df
```

```python
def transform_data(df, output_file):
    result_list = []
    
    start_time = time.time()

    for index, row in df.iterrows():
        id = row['id']
        question = row['input']
        correct_answer = row['options'][row['gold_index']]
        
        entry = {
            "id": id,
            "Question": question,
            "Answer": correct_answer,
            "Class_ID": row['class_id']
        }
        result_list.append(entry)
    
    with open(output_file, "w") as json_file:
        json.dump(result_list, json_file, indent=4)
    
    end_time = time.time()
    total_time = end_time - start_time
    
    total_pairs = len(result_list)
    
    print(f"Data transformation completed. {total_pairs} records saved to {output_file}")
    print(f"Total question-answer pairs extracted: {total_pairs}")
    print(f"Time taken for processing: {total_time:.2f} seconds")
```

```python
if __name__ == "__main__":
    output_file = "transformed_data.json"
    transform_data(df, output_file)
```
